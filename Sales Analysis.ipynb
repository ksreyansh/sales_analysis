### RFM Analysis

```python
data.head()
```

```python
m = data.groupby('Client')['Total'].sum()
m = pd.DataFrame(m).reset_index()
```

```python
m.head()
```

```python
#Frequency
f = data.groupby('Client')['Order_number'].count()
f = f.reset_index()
f.columns = ['Client', 'Frequency']
```

```python
f.head()
```

```python
#Recency
#Part 1
#Extracting the difference between the maximum date of purchase and last date of purchase
last_day = max(data['Sale_datetime'])
data['difference'] = last_day - data['Sale_datetime']
```

```python
data.sample()
```

```python
def get_Days(x):
    y = str(x).split()[0]
    return int(y)
```

```python
data['difference'] = data['difference'].apply(get_Days)
```

```python
data.head()
```

```python
#Part 2
r = data.groupby('Client')['difference'].min()
r = r.reset_index()
r.columns = ['Client', 'Recency']
```

```python
r.head()
```

```python
#Merge r, f abd m dataframes
group_df = pd.merge(m, f, on='Client', how='inner')
RFM_data = pd.merge(group_df, r, on='Client', how='inner')
RFM_data.columns = ['Client', 'Monetary', 'Frequency', 'Recency']
```

```python
RFM_data.head()
```

```python
RFM_data["FrequencyScore"] = pd.cut(RFM_data["Frequency"], 
                                       bins=[-1, 
                                         np.percentile(RFM_data["Frequency"], 25), 
                                             np.percentile(RFM_data["Frequency"], 50), 
                                             np.percentile(RFM_data["Frequency"], 75), 
                                             RFM_data["Frequency"].max()], 
                                       labels=[1, 2, 3, 4]).astype("int")
```

```python
RFM_data["RecencyScore"] = pd.cut(RFM_data["Recency"], 
                                       bins=[-1, 
                                         np.percentile(RFM_data["Recency"], 25), 
                                             np.percentile(RFM_data["Recency"], 50), 
                                             np.percentile(RFM_data["Recency"], 75), 
                                             RFM_data["Recency"].max()], 
                                       labels=[1, 2, 3, 4]).astype("int")
```

```python
RFM_data["MonetaryScore"] = pd.cut(RFM_data["Monetary"], 
                                       bins=[-1, 
                                         np.percentile(RFM_data["Monetary"], 25), 
                                             np.percentile(RFM_data["Monetary"], 50), 
                                             np.percentile(RFM_data["Monetary"], 75), 
                                             RFM_data["Monetary"].max()], 
                                       labels=[1, 2, 3, 4]).astype("int")
```

```python
RFM_data.head()
```

```python
RFM_data['Total_score'] = RFM_data['FrequencyScore'] + RFM_data['RecencyScore'] + RFM_data['MonetaryScore']
```

## Customer Segmentation Report

#### 1. High engagement and high value customers

```python
RFM_data.loc[((RFM_data['FrequencyScore'] == 4) & (RFM_data['MonetaryScore'] == 4) & (RFM_data['RecencyScore'] == 2))].count()[0]
```

There are 464 High engagement and High Value customers. These are the most valuable customers for the business. They made more than 9k in sales amount and more than 6 purchases within 4 months since the most recent date of purchase

#### 2. High Engagement and low value customers

```python
RFM_data.loc[((RFM_data['FrequencyScore'] == 4) & (RFM_data['MonetaryScore'] == 4) & (RFM_data['RecencyScore'] == 3))].count()[0]
```

There are 273 High Engagement and Low value customers. These customers contributed less than 9k but more than 2k in total amount

#### 3. Medium engagement and medium value customers

```python
RFM_data.loc[((RFM_data['FrequencyScore'] == 3) & (RFM_data['MonetaryScore'] == 3) & (RFM_data['RecencyScore'] == 3))].count()[0]
```

There are 273 High Engagement and Low value customers. These customers contributed less than 9k but more than 2k in total amount but purchased less than 50 times

#### 4. Low Engagement and low value customers

```python
RFM_data.loc[((RFM_data['FrequencyScore'] == 2) & (RFM_data['MonetaryScore'] == 1))].count()[0]
```

There are 284 low engagement and low value customers. These customers might churn out at any time

```python
RFM_data.loc[((RFM_data['FrequencyScore'] == 1) & (RFM_data['MonetaryScore'] == 1) & (RFM_data['RecencyScore'] == 4))].count()[0]
```
